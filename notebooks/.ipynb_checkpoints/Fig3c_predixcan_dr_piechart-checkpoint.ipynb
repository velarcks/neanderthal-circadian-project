Proportion of circadian genes that contain evidence of regulatory differences between archaics and humans from splice-altering variants (SAV) and divergent regulation (DR) predictions.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# LOAD DATA
all_circadian = pd.read_csv('../data/circadian_genes.list', sep='\t')
dr = pd.read_csv('data/predixcan_dr_circadian.txt', sep='\t')[['GeneID','GeneName']].drop_duplicates()
sav = pd.read_csv('data/alt_splicing_circadian.txt', sep='\t')[['GeneID','GeneName']].drop_duplicates()
```

```python
labels = ['CIRCADIAN', 'DR', 'SAV']
data = [len(all_circadian)-len(dr)-len(sav), len(dr), len(sav)]
```

```python
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int((pct / 100)*sum(allvalues))
    return "{:.1f}%\n({:.0f})".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: "{:.1f}%\n({:.0f})".format(pct, (pct * sum(data) / 100)),
                                  labels = labels,
                                  shadow = False,
                                  startangle = 90,
                                  colors = ['skyblue', 'beige', 'ivory'],
                                  textprops={'fontsize': 15}
                                  #textprops = dict(color ="magenta")
                                 )

#plt.savefig('plots/predixcan_dr_sav_piechart.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

Plot the cumulative fraction of introgressed circadian loci that increase morningness propensity. The Morning/evening person trait data was extracted from the UKBiobank GWAS by the Neale Lab.

```python
# ---------------------------------------------------------------
```

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pylab as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
# -----------------------------------------------------------------------
# PLOT % MORNINGNESS AFTER LD CLUMPING THE MORNINGNESS VARIANTS BY R2 0.9 CEU
# ALL INTROGRESSED MORNING/EVENING PERSON ASSOCIATIONS
```

```python
clump_plot = pd.read_csv('data/plotting_ukbb_morningness_introgressed_direction_of_effect_clump.txt',
                            sep='\t')
```

```python
# -----------------------------------------------------------------------
# PLOT % MORNINGNESS AFTER LD CLUMPING THE MORNINGNESS VARIANTS BY R2 0.9 CEU
# MORNING/EVENING PERSON ASSOCIATIONS IN CIRCADIAN GENES
```

```python
clump_c_plot = pd.read_csv('data/plotting_ukbb_morningness_introgressed_circadian_direction_of_effect_clump.txt',
                            sep='\t')
```

```python
# -----------------------------------------------------------------------
# PLOT % MORNINGNESS
```

```python
fig, ax = plt.subplots()

sns.lineplot(x='log10 P-Value',y='% Morningness',data=clump_plot, ax=ax, ci=False,
             color='black', linewidth=2.5, marker='o',markersize=6, alpha=0.5, label="Morning person associations")
sns.lineplot(x='log10 P-Value',y='% Morningness',data=clump_c_plot, ax=ax, ci=False,
             color='blue', linewidth=2.5, marker='^',markersize=8, alpha=0.5, label="Morning person associations in circadian genes")
plt.xticks(rotation=90);
plt.legend(loc='lower left',fontsize='15')
plt.tight_layout()
#plt.savefig("plots/ukbb_morningness_introgressed_direction_of_effect_clump_both.pdf", dpi=300)
```

```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")
import matplotlib.pylab as plt
pd.set_option('display.max_rows', 600)
import numpy as np
import math
```

```python
# LOAD DATA generated in generate_nealelab_1180_circadian_input.ipynb
morningness = pd.read_csv('data/plotting_ukbb_morningness_direction_of_effect_circadian.txt',
                           sep='\t')
```

```python
morningness
```

```python
morningness['P-Value Range'] = pd.qcut(morningness['P-Value'], 6) #labels=['8.25e-15 to 1.78e-05',
                #'1.78e-05 to 1.81e-03','1.81e-03 to 8.42e-3','8.42e-3 to 2.24e-02','2.24e-02 to 0.05'])

sns.set(rc = {'figure.figsize':(15,8)})
sns.violinplot(x='P-Value Range',y='% Morningness',data=morningness)
plt.xticks(rotation=90)
plt.title('{} bins'.format(6))
```

```python
morningness['P-Value Range'] = pd.qcut(morningness['P-Value'], 10) #labels=['8.25e-15 to 1.78e-05',
                #'1.78e-05 to 1.81e-03','1.81e-03 to 8.42e-3','8.42e-3 to 2.24e-02','2.24e-02 to 0.05'])

sns.set(rc = {'figure.figsize':(15,8)})
sns.boxplot(x='P-Value Range',y='% Morningness',data=morningness)
plt.xticks(rotation=90)
plt.title('{} bins'.format(10))
```

```python
#sns.set(rc = {'figure.figsize':(15,8)})
sns.lineplot(x='log10 P-Value',y='% Morningness',data=morningness, color='blue', linewidth=2.5)
plt.xticks(rotation=90);
#plt.title('{} bins'.format(6))
plt.savefig("ukbb_morningness_introgressed_direction_of_effect_circadian.pdf", dpi=300)
```
